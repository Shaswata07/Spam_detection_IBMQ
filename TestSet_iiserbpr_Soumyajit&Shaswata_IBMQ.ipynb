```python
#the values of theta, phi and lambda from the training data set
ft=0
fp=pi/6
fl=pi/3
```

```python
#===================================================E1--00=======================================================#

#qcE1 is the fist example

q1 = QuantumRegister(2)
c1 = ClassicalRegister(2)
qcE1= QuantumCircuit(q1,c1)
#input state

qcE1.barrier()

qcE1.u3(ft,fp,fl,q1[0])
qcE1.barrier(q1)
qcE1.measure(q1,c1)

jobE1=execute(qcE1, backend=deviceV, shots=8192)
resultE1=jobE1.result()
countsE1=resultE1.get_counts(qcE1)

```

```python
qcE1.draw(output='mpl')
```

```python
plot_histogram(countsE1)
```
